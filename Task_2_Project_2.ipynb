# **Task-1**
### **Data Import and Cleaning**
**Objective:**

Prepare the data for analysis by ensuring its consistency and quality.

**Steps:**


*   Import a CSV file containing production data (e.g., temperature, catalyst concentration, reaction time).
*   Handle missing values by imputing or removing them.
Clean the data by eliminating invalid entries or outliers.





```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Import CSV file
data = pd.read_csv('/content/drive/MyDrive/Abdullah_Al_Mahfuz_Basic_Polymer_Production_Rate_Estimation.csv')
# Display the first few rows of the data
print(data.head())
# Handle missing values by removing rows with missing data
data.dropna(inplace=True)
# Remove outliers using the IQR method
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
data = data[~((data < (Q1 - 1.5 * IQR)) |(data > (Q3 + 1.5 * IQR))).any(axis=1)]
print(data.describe())
```

# **Task-2**
### **Simple Production Rate Calculation**
**Objective:**

Estimate the polymer production rate using a simplified model.

**Model:**

Production Rate = a × Temperature + b × Catalyst Concentration + c × Reaction Time

Here,

a,
b and
c are constants derived from prior data.

**Steps:**



*   Use the cleaned data to calculate the production rate based on the formula.
*   Ensure the calculation is efficient and handles edge cases (e.g., extreme values).






```python
# Define constants
a = 0.5
b = 1.2
c = 0.8
# Calculate the production rate
data['Polymer_Production_Rate (kg/h)'] = a * data['Temperature (°C)'] + b * data['Catalyst_Conc (%)'] + c * data['Reaction_Time (h)']
# Display the first few rows with the new Production Rate column
print(data.head())
```

# **Task-3**
## **Visualization**
### **Objective:**

Generate visual insights into the relationships between variables.

### **Graphs:**
**Temperature vs. Production Rate:**

A line plot showing how temperature influences production rates.

**Bar Chart for Catalyst Concentrations:**

A bar chart displaying production rates for different levels of catalyst concentration.

```python
# Plot the relationship between temperature and production rate
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Temperature (°C)', y='Polymer_Production_Rate (kg/h)', data=data)
plt.title('Temperature vs Production Rate')
plt.xlabel('Temperature')
plt.ylabel('Production Rate')
plt.show()
```

# **Temperature vs. Production Rate**
This graph presents a line plot that highlights the relationship between temperature and the polymer production rate. The trend illustrates how temperature influences the production rate, showing whether the reaction benefits from higher temperatures or reaches a saturation point. **For example:**

**A positive slope** would indicate that increasing the temperature accelerates the reaction and boosts production.
Any non-linear trends (e.g., plateaus or declines) could point to optimal temperature ranges or thermal degradation effects at higher levels.
This visualization helps identify the temperature range where polymer production is maximized without adversely affecting the process. It provides insights into the thermal stability and efficiency of the reaction, making it a crucial tool for fine-tuning reaction parameters.

```python
# Plot bar chart showing polymer production rates for different catalyst concentrations
plt.figure(figsize=(10, 6))
sns.barplot(x='Catalyst_Conc (%)', y='Polymer_Production_Rate (kg/h)', data=data)
plt.title('Catalyst Concentration vs Production Rate')
plt.xlabel('Catalyst Concentration')
plt.ylabel('Production Rate')
plt.show()
```

# **Bar Chart for Catalyst Concentrations**
The bar chart compares polymer production rates across varying catalyst concentrations. Each bar represents a specific catalyst concentration, and its height corresponds to the respective production rate. Key insights include:

Identifying the catalyst concentration that results in the highest production rate.
Detecting diminishing returns, where increasing the concentration further does not proportionally increase production.
Spotting any potential inefficiencies, such as very low production rates at certain concentrations.
By examining this chart, users can understand the impact of catalyst concentration on the reaction's efficiency and determine the most cost-effective concentration for optimizing production.

# **Full Example Code**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Import CSV file
data = pd.read_csv('/content/drive/MyDrive/Abdullah_Al_Mahfuz_Basic_Polymer_Production_Rate_Estimation.csv')

# Display the first few rows of the data
print(data.head())

# Handle missing values by removing rows with missing data
data.dropna(inplace=True)

# Remove outliers using the IQR method
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
data = data[~((data < (Q1 - 1.5 * IQR)) |(data > (Q3 + 1.5 * IQR))).any(axis=1)]

print(data.describe())

# Define constants
a = 0.5
b = 1.2
c = 0.8

# Calculate the production rate
data['Polymer_Production_Rate (kg/h)'] = a * data['Temperature (°C)'] + b * data['Catalyst_Conc (%)'] + c * data['Reaction_Time (h)']

# Display the first few rows with the new Production Rate column
print(data.head())

# Plot the relationship between temperature and production rate
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Temperature (°C)', y='Polymer_Production_Rate (kg/h)', data=data)
plt.title('Temperature vs Production Rate')
plt.xlabel('Temperature')
plt.ylabel('Production Rate')
plt.show()

# Plot bar chart showing polymer production rates for different catalyst concentrations
plt.figure(figsize=(10, 6))
sns.barplot(x='Catalyst_Conc (%)', y='Polymer_Production_Rate (kg/h)', data=data)
plt.title('Catalyst Concentration vs Production Rate')
plt.xlabel('Catalyst Concentration')
plt.ylabel('Production Rate')
plt.show()
```
